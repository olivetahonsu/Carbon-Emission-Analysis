```python
# This project is based on the Individual Carbon Footprint Calculation data downloaded from https://www.kaggle.com/

# The aim of the project is to reveal how individual lifestyle can impact their carbon footprint.
```

```python
# Import the necesary Python libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data set as a dataframe called 'carbon_footprint'
carbon_footprint = pd.read_csv('Carbon Emission.csv')
```

```python
# Inspect the first 5 rows of the data.
carbon_footprint.head()
```

```python
# Inspect the last 5 rows of the dataset.
carbon_footprint.tail()
```

```python
#Group average individual carbon emission by 'Heating Energy Source' and convert the grouped data to a dataframe.
grouped_data = carbon_footprint.groupby('Heating Energy Source')['CarbonEmission'].mean().reset_index()
```

```python
#The grouped data 
print(grouped_data)
```

```python
#Plot the grouped data.
bar = sns.barplot(data = grouped_data, x = 'Heating Energy Source', y = 'CarbonEmission')
bar.set_title('Mean Carbon Emission by Heating Energy Source')
plt.xlabel('Heat Energy Source')
plt.ylabel('Mean Carbon Emission')
plt.show()
```

```python
#Group average individual carbon emission by 'Frequency of Air Travel' and convert the grouped data to a dataframe.
grouped_data = carbon_footprint.groupby('Frequency of Traveling by Air')['CarbonEmission'].mean().reset_index()

print(grouped_data)
```

```python
#Plot the grouped data.
bar = sns.barplot(data = grouped_data, x = 'Frequency of Traveling by Air', y = 'CarbonEmission')
bar.set_title('Mean Carbon Emission by Frequency of Air Travel')
plt.xlabel('Frequency of Air Travel')
plt.ylabel('Mean Carbon Emission')
plt.show()
```

```python
#Group average individual carbon emission by 'Waste Bag Weekly Count' and convert the grouped data to a dataframe.
grouped_data = carbon_footprint.groupby('Waste Bag Weekly Count')['CarbonEmission'].mean().reset_index()
```

```python
#The grouped data.
print(grouped_data)
```

```python
#Plot the grouped data.
bar = sns.barplot(data=grouped_data, x='Waste Bag Weekly Count', y='CarbonEmission')
bar.set_title('Mean Carbon Emission by Waste Bag Weekly Count')
plt.xlabel('Waste Bag Weekly Count')
plt.ylabel('Average Carbon Emission')
plt.figure(figsize=(10, 6))  # This line sets the size of the figure, but it's not directly related to sns.barplot()
plt.show()
```

```python
#Group average individual carbon emission by 'Transport type' and convert the grouped data to a dataframe.
grouped_data = carbon_footprint.groupby('Transport')['CarbonEmission'].mean().reset_index()
```

```python
#Group average individual carbon emission by 'Vehicle Type' and convert the grouped data to a dataframe.
grouped_data = carbon_footprint.groupby('Vehicle Type')['CarbonEmission'].mean().reset_index()
```

```python
sizes = grouped_data['CarbonEmission']
labels = grouped_data['Vehicle Type']

# Set seaborn style
sns.set_style("whitegrid")

# Create a pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=0)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

plot_title = 'Probability Density Functions'

#plt.title(plot_title, fontsize=16, color='blue')

# Show the plot
plt.show()
```
